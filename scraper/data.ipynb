Something to note is that the maximum number of items that can be returned in a search is 1000, so 950 is the largest offset that still works. I'm pulling exactly 1000 from every year since 2000. This means that my dataset will be comprised of 22000 playlists, each of which contains songs and the characteristic song features. An example of one such pull is below, with data cleaned and in a usable dataframe format.

```python
#Get songs for every year from 2000-2022

artist_name = []
track_name = []
popularity = []
track_id = []

for i in range(0,950,50):
        track_results = sp.search(q='year:2000', type='track', limit=50,offset=i)
        for i, t in enumerate(track_results['tracks']['items']):
            artist_name.append(t['artists'][0]['name'])
            track_name.append(t['name'])
            track_id.append(t['id'])
            popularity.append(t['popularity'])

track_dataframe = pd.DataFrame({'artist_name' : artist_name, 'track_name' : track_name, 'track_id' : track_id, 'popularity' : popularity})
print(track_dataframe.shape)
track_dataframe.head()
res = track_dataframe.to_json()
res
```
